```python
problem_logs_sorted = problem_logs.sort_values(['student_id', 'start_time'])
problem_logs_sorted.head(10)
```

```python
problem_logs_sorted['next_correct'] = (problem_logs_sorted.groupby('student_id')['correct'].shift(-1))
data_with_target = problem_logs_sorted.dropna(subset=['next_correct'])
data_with_target.head()
```

```python
merged_data = data_with_target.merge(problem_details,on='problem_id', how='left')
merged_data = merged_data.merge(student_details,on='student_id',how='left')
```

```python
print(merged_data.columns)
merged_data.shape
```

```python
import numpy as np

student_problem_counts = merged_data['student_id'].value_counts() 
eligible_students = student_problem_counts[
    (student_problem_counts >= 5) & (student_problem_counts <= 20)
].index

np.random.seed(42)

train_students = np.random.choice(eligible_students, size=10000, replace=False)
remaining_students = np.setdiff1d(eligible_students, train_students)
test_students = np.random.choice(remaining_students, size=1000, replace=False)

train_data = merged_data[merged_data['student_id'].isin(train_students)]
test_data = merged_data[merged_data['student_id'].isin(test_students)]

print(f"Students for train data {train_data['student_id'].nunique()} | Row: {train_data.shape[0]}")
print(f"Students for test data {test_data['student_id'].nunique()} | Row: {test_data.shape[0]}")
```

```python
os.makedirs("data",exist_ok=True)
train_data.to_csv("data/train_data.csv",index=False)
test_data.to_csv("data/test_data.csv",index=False)
```
